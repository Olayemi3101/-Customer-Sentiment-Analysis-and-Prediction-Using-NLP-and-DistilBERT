#### Business Introduction
ShopEase Europe is a large online shopping company that sells different products across Europe.
Every day, customers leave reviews, comments, and complaints on the website, social media, and customer support channels. With over a million users each month, there is too much feedback for staff to read manually.
The company wants to understand what customers are saying and how they feel. To maintain high customer satisfaction and improve service delivery.

#### Problem Statement
- Thousands of reviews are received every day, making it difficult for employees to read and analyze them all.
- Different analysts may interpret the same review differently, leading to inconsistent results.
- Although feedback is available, it takes too much time to turn it into useful business actions.

#### Project Objectives

- Clean and prepare customer reviews written in different languages.
- Build machine learning models that can predict customer sentiment.
- Identify the main reasons customers are happy or unhappy.
- Create summaries of customer opinions.
- Display results in an interactive dashboard.

##### Language distribution within customer reviews.

- Detect the Language of Each Review

```python
def detect_language(text):
    try:
        return detect(text)
    except LangDetectException:
        return "unknown"
```

```python
df_clean["language"] = df_clean["review"].apply(detect_language)
```

     Language detection revealed that customer reviews are primarily written in English (44%), French (20%), German (20%), and Spanish (15%). This confirms the multilingual nature of ShopEase's customer base and supports the need for multilingual NLP techniques for sentiment analysis.

##### NLP preprocessing workflow

- Lowercase text on 'reviews'

```python
df_clean["processed_review"] = df_clean["review"].str.lower()
```

- Remove URLs

```python
df_clean["processed_review"] = df_clean["processed_review"].apply(lambda x: re.sub(r"http\S+|www\S+", "", str(x)))
```

- Remove punctations and emojis

```python
df_clean["processed_review"] = df_clean["processed_review"].apply(lambda x: re.sub(r"[^\w\s]", "", str(x)))
```

- Remove stopwords in the top 4 languages

```python
# Create stopword sets for different languages
stop_en = set(stopwords.words("english"))
stop_fr = set(stopwords.words("french"))
stop_de = set(stopwords.words("german"))
stop_es = set(stopwords.words("spanish"))

# Function to remove stopwords based on language
def remove_stopwords(text, lang):

    text = str(text).lower()

    words = text.split()

    if lang == "en":
        stop_words = stop_en
    elif lang == "fr":
        stop_words = stop_fr
    elif lang == "de":
        stop_words = stop_de
    elif lang == "es":
        stop_words = stop_es
    else:
        return text

    filtered_words = [word for word in words if word not in stop_words]

    return " ".join(filtered_words)

# Apply the remove_stopwords function to the processed_review column based on the detected language
df_clean["processed_review"] = df_clean.apply(lambda row: remove_stopwords(row["processed_review"], row["language"]), axis=1)
```

```python
df_clean[["review", "processed_review", "language"]].head(10)
```

- Lemmatization of the 4 languages 

```python
#Load language models

nlp_en = spacy.load("en_core_web_sm")
nlp_fr = spacy.load("fr_core_news_sm")
nlp_de = spacy.load("de_core_news_sm")
nlp_es = spacy.load("es_core_news_sm")

#lemmatization function
def lemmatize_text(text, lang):
    text = str(text).lower()

    if lang == "en":
        doc = nlp_en(text)
    elif lang == "fr":
        doc = nlp_fr(text)
    elif lang == "de":
        doc = nlp_de(text)
    elif lang == "es":
        doc = nlp_es(text)
    else:
        return text

    lemmas = [token.lemma_ for token in doc if token.is_alpha]

    return " ".join(lemmas)

# Apply lemmatization to the processed reviews
df_clean["processed_review"] = df_clean.apply(
    lambda row: lemmatize_text(row["processed_review"], row["language"]), axis=1)
```
