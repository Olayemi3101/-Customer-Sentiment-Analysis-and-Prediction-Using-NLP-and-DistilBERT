##### Load the dataset

```python
df_clean = pd.read_csv(r'C:\Users\balog\OneDrive\Documents\Olayemi\Intership\Customer Feedback Analysis\Data\cleaned_amazon_reviews.csv')
```

##### Dashboard Reporting Tables

- Category Summary

```python
category_summary = (
    df_clean.groupby("product_category").agg(Total_Reviews=("review_id","count"),Average_Rating=("rating","mean")).reset_index())

category_summary
```

```python
category_summary.to_excel("category_summary.xlsx", index=False)
```

- Sentiment Summary

```python
sentiment_summary = (df_clean.groupby(["product_category","sentiment"]) .size().reset_index(name="total_reviews"))
sentiment_summary
```

```python
sentiment_summary.to_excel("sentiment_summary.xlsx", index=False)
```

- Country Summary

```python
country_summary = (df_clean.groupby("country").agg(Total_Reviews=("review_id","count"), Average_Rating=("rating","mean")).reset_index())
country_summary
```

```python
country_summary.to_excel("country_summary.xlsx", index=False)
```
