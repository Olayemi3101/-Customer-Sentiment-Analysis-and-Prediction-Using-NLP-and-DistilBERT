#### Business Introduction
ShopEase Europe is a large online shopping company that sells different products across Europe.
Every day, customers leave reviews, comments, and complaints on the website, social media, and customer support channels. With over a million users each month, there is too much feedback for staff to read manually.
The company wants to understand what customers are saying and how they feel. To maintain high customer satisfaction and improve service delivery.

#### Problem Statement
- Thousands of reviews are received every day, making it difficult for employees to read and analyze them all.
- Different analysts may interpret the same review differently, leading to inconsistent results.
- Although feedback is available, it takes too much time to turn it into useful business actions.

#### Project Objectives

- Clean and prepare customer reviews written in different languages.
- Build machine learning models that can predict customer sentiment.
- Identify the main reasons customers are happy or unhappy.
- Create summaries of customer opinions.
- Display results in an interactive dashboard.
